# Data Exploration

**Distribution Analysis: Streaming Durations**

```python
from logging import log
plt.figure(figsize=(12, 6))
sns.histplot(clean_data['ms_played'], bins=50, kde=True)
plt.title('Distribution of Streaming Durations')
plt.xlabel('Milliseconds Played')
plt.ylabel('Frequency')
plt.show()
```

**Distribution Analysis: Platforms**

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='platform', data=clean_data)
plt.title('Distribution of Streaming Platforms')
plt.xlabel('Platform')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

**Distribution Analysis: Countries**

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='conn_country', data=clean_data)
plt.title('Distribution of Streaming Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

**Identifying: Top 10 Tracks**

```python
top_tracks = clean_data['master_metadata_track_name'].value_counts().head(10)
print("Top 10 Tracks:\n", top_tracks)
```

**Identifying: Top 10 Artists**

```python
top_artists = clean_data['master_metadata_album_artist_name'].value_counts().head(10)
print("\nTop 10 Artists:\n", top_artists)
```

**Identifying: Top 10 Albums**

```python
top_albums = clean_data['master_metadata_album_album_name'].value_counts().head(10)
print("\nTop 10 Albums:\n", top_albums)
```

**Visualize Top Tracks, Artists, and Albums**

```python
# Visualize top tracks
plt.figure(figsize=(14, 6))
top_tracks.plot(kind='bar')
plt.title('Top 10 Most Streamed Tracks')
plt.xlabel('Track')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize top artists
plt.figure(figsize=(14, 6))
top_artists.plot(kind='bar')
plt.title('Top 10 Most Streamed Artists')
plt.xlabel('Artist')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize top albums
plt.figure(figsize=(14, 6))
top_albums.plot(kind='bar')
plt.title('Top 10 Most Streamed Albums')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

**Analyzing Streaming Habits Over Time**

```python
# Convert 'ts' column to datetime format
clean_data['ts'] = pd.to_datetime(clean_data['ts'])

# Set 'ts' column as the index
clean_data.set_index('ts', inplace=True)

# Resample the data on a monthly basis
monthly_streaming_trends = clean_data.resample('M')['ms_played'].sum()

# Plot streaming trends
plt.figure(figsize=(14, 6))
monthly_streaming_trends.plot(marker='o')
plt.title('Monthly Streaming Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Milliseconds Played')
plt.show()
```

**Summarize Key Findings**

```python
# Display summary statistics of the streaming duration
summary_statistics = clean_data['ms_played'].describe()
print("Summary Statistics of Streaming Duration:\n", summary_statistics)
```

```python
# Identify the most common platform used for streaming
most_common_platform = clean_data['platform'].mode().values[0]

# Identify the country with the highest total streaming duration
top_country = clean_data.groupby('conn_country')['ms_played'].sum().idxmax()

# Display insights
print("Most Common Platform: ", most_common_platform)
print("Country with the Highest Total Streaming Duration: ", top_country)
```
